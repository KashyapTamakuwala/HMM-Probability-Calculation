```python
ini=[0.6,0.4]
A=[[0.7,0.3],[0.4,0.6]]
b=[[0.1,0.4,0.5],[0.7,0.2,0.1]]

lisofseq=['sHHHH','sHHHC', 'sHHCH', 'sHCHH', 'sCHHH','sHHCC', 'sHCCH', 'sCCHH', 'sHCHC', 'sCHCH', 'sCHHC','sHCCC', 'sCHCC','sCCHC', 'sCCCH','sCCCC']
listofpro=[] 
j=[0,1,0,2]

for s in lisofseq:
    pro=1
    for i in range(0,len(s)-1):
        if(s[i]=='s'):
            if(s[i+1]=='H'):
                pro=pro*ini[0]
                if(j[i]==0):
                    pro=pro*b[0][0]
                elif(j[i]==1):
                    pro=pro*b[0][1]
                else:
                    pro=pro*b[0][2]
                continue
            else:
                pro=pro*ini[1]
                if(j[i]==0):
                    pro=pro*b[1][0]
                elif(j[i]==1):
                    pro=pro*b[1][1]
                else:
                    pro=pro*b[1][2]
                continue
        if(s[i]=='H'):
            if(s[i+1]=='H'):
                pro=pro*A[0][0]
                if(j[i]==0):
                    pro=pro*b[0][0]
                elif(j[i]==1):
                    pro=pro*b[0][1]
                else:
                    pro=pro*b[0][2]
                continue
            else:
                pro=pro*A[0][1]
                if(j[i]==0):
                    pro=pro*b[1][0]
                elif(j[i]==1):
                    pro=pro*b[1][1]
                else:
                    pro=pro*b[1][2]
                continue
        if(s[i]=='C'):
            if(s[i+1]=='H'):
                pro=pro*A[1][0]
                if(j[i]==0):
                    pro=pro*b[0][0]
                elif(j[i]==1):
                    pro=pro*b[0][1]
                else:
                    pro=pro*b[0][2]
                continue
            else:
                pro=pro*A[1][1]
                if(j[i]==0):
                    pro=pro*b[1][0]
                elif(j[i]==1):
                    pro=pro*b[1][1]
                else:
                    pro=pro*b[1][2]
                continue
    listofpro.append(pro)
         
    
        
```

```python
data=[]
for i in range(0,len(lisofseq)):
    temp=[lisofseq[i],listofpro[i]]
    data.append(temp)
```
